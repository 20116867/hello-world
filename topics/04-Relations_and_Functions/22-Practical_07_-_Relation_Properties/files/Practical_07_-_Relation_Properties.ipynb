```python
# Question 4

def vowels(word):
    """Return the set of vowels in word."""
    return {c for c in word if c in 'aeiou'}

vowels("fire")
A = {"eject", "table", "green", "light", "bread", "heart", "house", "beach"}
R = {(a, b) for a in A for b in A if vowels(a) == vowels(b)}
print(R)
print(len(R))
draw_relation_as_venn_diagram(A, A, R)

```
